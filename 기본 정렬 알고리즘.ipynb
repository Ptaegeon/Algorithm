```python
#Selection Sort

def selection(arr):
  
  n = len(arr)
  for i in range(n-1):
    min = i

    for x in range(i, n):
      if arr[x]<arr[min]:
        min = x
    
    arr[i], arr[min] = arr[min], arr[i]
  
  print(arr)
  return 

```

```python
arr = [8,7,3,4,5,1]
selection(arr)
```

```python
#Bubble sort

def Bubble(arr):

  for i in range(len(arr)):

    for j in range(len(arr)-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return print (arr)
```

```python
arr = [8,7,3,4,5,1]
Bubble(arr)
```

```python
#quick sort // stable but not in-place

def quick_stable(A):

  if len(A) <=1 : return A

  S, M, L = [], [], []
  pivot = A[0]
  left = 0
  right = len(A)-1

  while left <= right :
    if A[left] < pivot:
      S.append(A[left])
    elif A[left] > pivot:
      L.append(A[left])
    else : M.append(A[left])

    left+= 1
  
  return quick_stable(S) + M + quick_stable(L)
```

```python
arr = [8,7,3,4,5,1]
quick_stable(arr)
```

```python
#quick sort // stable but not in-place

def quick_inplace(A, first, last):

  if first >= last : return 

  pivot = A[first]
  left = first+1
  right = last

  while left <= right :
    
    while left<=last and pivot > A[left]:
      left += 1
    while  pivot < A[right]: 
      right -= 1
    if left<= right:
      A[left], A[right] = A[right], A[left]
  
  A[right], A[first]= A[first], A[right]
  
  quick_inplace(A, first, right-1)
  quick_inplace(A, left, last)   
  return 

```

```python
arr = [8,7,3,4,5,1]
quick_inplace(arr, 0, len(arr)-1)
print(arr)
```

```python
#Merge sort

def Mergesort(A, first, last):

  if first >= last : return

  mid = (first+last)//2

  Mergesort(A, first, mid)
  Mergesort(A, mid+1, last)
  Merge_two_sorted_list(A,first,last)

  return

def Merge_two_sorted_list(A,first, last):

  B = []
  mid = (first+last)//2
  i, j = first, mid+1

  while i <= mid and j<= last :
    
    if A[i]<= A[j] :
      B.append(A[i])
      i+= 1
    else :
      B.append(A[j])
      j+= 1
  
  for k in range(i, mid+1) :
    B.append(A[k])

  for k in range(j, last+1) :
    B.append(A[k])

  for i in range(first, last+1):
    A[i] = B[i-first]
  return 
```

```python
#@title
#heap sort

def heapsort(arr) :
  
  for k in range(len(arr)-1, -1, -1) :
   heapfy_down(arr,k)
  return arr

def heapfy_down(arr, index):

  if 2*index +1 >= len(arr) : return
  
  while 2*index +2 <= len(arr)-1 or 2*index +1 <= len(arr)-1  :  
    parent = index
    left = index *2 + 1
    right = index *2 + 2
    
    if len(arr)-1  < right:
      if arr[parent] < arr[left]:
        if len(arr)-1 < right: 
            arr[parent],  arr[left] = arr[left], arr[parent]
            index = left          
      
    else : 
      if len(arr)-1 <=  right :
        if arr[right]> arr[left]:
          arr[parent],  arr[right] = arr[right], arr[parent]
          index= right   
        if arr[right] <= arr[left]:
          arr[parent],  arr[left] = arr[left], arr[parent]
          index= left      
     
  return
```
