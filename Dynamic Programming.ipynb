```python
def fib1(n):

  if n==1 or n==0 : return n

  return fib1(n-1) + fib1(n-2)

class Fib :
  def __init__(self) :
    self.size = 1
    self.memo = [0,1]
    return 

  def fib(self,n):

    if n<= 0 : return

    if self.size >= n:
      return self.memo[n] 
    
    for k in range(self.size+1, n+1):
      self.memo.append(self.memo[k-1] + self.memo[k-2])
      self.size +=1
    return self.memo[n]
  
```

```python
import time
N=10
start = time.time()
fib1(N)
print((time.time()-start)*1000, "ms")

fib = Fib()
start = time.time()
fib.fib(N)
print((time.time()-start)*1000, "ms")

```

```python
#LCS  memoization

def lcs_recur(X,Y,m,n):

  if m==0 or n==0 : return 0

  elif X[m-1] == Y[n-1]:
    return lcs_recur(X,Y,m-1,n-1) +1
  
  else : 
    return max(lcs_recur(X,Y,m,n-1),
               lcs_recur(X,Y,m-1,n))
```

```python
#LCS Tabulation

def lcs_dp(X,Y):

  m = len(X)
  n = len(Y) 

  L= [[None]*(n+1) for _ in range(m+1)]

  for i in range(m+1) :
    for j in range(n+1):
      if i ==0 or j==0 :
        L[i][j] = 0
      
      elif X[i-1] == Y[j-1] : 
        L[i][j] = L[i-1][j-1] +1
      
      else :
        L[i][j] = max(L[i-1][j], L[i][j-1])

  return L[m][n]
```

```python
X="ababbbaaa"
Y='abaabbbb'
print(lcs_recur(X,Y,len(X),len(Y)), lcs_dp(X,Y))
```
